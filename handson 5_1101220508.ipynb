```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
#mengimport packages yang diperlukan
```

```python
xz = pd.read_csv("FuelConsumptionCo2.csv")
xz.head()
#read data
```

```python
xz.describe()
```

```python
cxz = xz[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cxz.head(10)
```

```python
viz = cxz[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

```python
plt.scatter(cxz.FUELCONSUMPTION_COMB, cxz.CO2EMISSIONS, color='maroon')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cxz.ENGINESIZE, cxz.CO2EMISSIONS, color='green')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
msk = np.random.rand(len(xz)) < 0.8
train = cxz[msk]
test = cxz[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)
# The coefficients
print('Coefficients: ', regr.coef_) 
print('Intercept: ', regr.intercept_) 
```

```python
plt.scatter(train["ENGINESIZE"], train["CO2EMISSIONS"], color='blue', label="Data asli")
plt.plot(train["ENGINESIZE"], regr.coef_[0] * train["ENGINESIZE"] + regr.intercept_, '-r', label="Regresi linier")
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.legend()
plt.show()
```
